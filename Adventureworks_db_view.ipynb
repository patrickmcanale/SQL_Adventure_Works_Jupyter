```python
### Finding how many employees were listed more than once

pd.read_sql_query("""
SELECT COUNT(*)
FROM
(SELECT BusinessEntityID, COUNT(BusinessEntityID) AS listed_twice
FROM AdventureWorks2016.dbo.EmployeeView
GROUP BY BusinessEntityID
HAVING COUNT(*) > 1) AS a;
""", connection)

```
